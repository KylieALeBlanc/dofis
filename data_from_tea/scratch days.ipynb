```python
import pandas as pd
import sys
sys.path.append("../")
import os
import engarde.decorators as ed
from library import clean_tea_schools
from library import clean_tea

data_path = '/Users/kylieleblancKylie/domino/dofis/data/'

days = pd.read_csv(os.path.join(data_path, 'tea', 'cdays', 'days_yr1718.csv'), sep=",")
vars_to_keep = {'DISTRICT': 'district', 'DISTNAME': 'distname', 
                   'CAMPUS': 'campus', 'CAMPNAME': 'campname',
                   'TRACK': 'track',
                   'TOTAL_DAYS': 'days'}
days = clean_tea.filter_and_rename_cols(days, vars_to_keep)
days.head()
```

```python
days = days.groupby(by = ['district', 'distname', 'campus', 'campname']).max().reset_index()
days = days[['district', 'distname', 'campus', 'campname', 'days']]
days.head()
```

```python
ddays = days.groupby(by = ['district', 'distname']).agg({'days': ['min', 'mean', 'max']}).reset_index()
ddays.columns = [' '.join(col).strip() for col in ddays.columns.values]
ddays.columns
ddays = ddays.rename({'days min': 'days_min', 'days mean': 'days_mean', 'days max': 'days_mean'}, axis = 'columns')
len(ddays)
```
