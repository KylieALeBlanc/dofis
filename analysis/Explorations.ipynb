# Exemptions by Implementation Year

```python
data = pd.read_csv((os.path.join(data_path, 'master_data_district.csv')))
data = data[data.year == 2018]
data.doi_year.value_counts()
```

## All groups

```python
stubnames = sorted(
    set([match[0] for match in data.columns.str.findall(
    r'reg.*').values if match != [] ])
    )

proportion = []
number = []
for reg in stubnames:
    proportion.append(data[reg].mean())
    number.append(data[reg].sum())

regs = pd.DataFrame(
    {'law': stubnames,
     'proportion': proportion,
     'number': number
    }).sort_values(by = ['proportion'], ascending = False)
regs.head(15)
```

```python
data_1stgroup = data[data.doi_year == 2017]
stubnames = sorted(
    set([match[0] for match in data_1stgroup.columns.str.findall(
    r'reg.*').values if match != [] ])
    )

proportion = []
number = []
for reg in stubnames:
    proportion.append(data_1stgroup[reg].mean())
    number.append(data_1stgroup[reg].sum())

regs = pd.DataFrame(
    {'law': stubnames,
     'proportion': proportion,
     'number': number
    }).sort_values(by = ['proportion'], ascending = False)
regs.head(15)
```
