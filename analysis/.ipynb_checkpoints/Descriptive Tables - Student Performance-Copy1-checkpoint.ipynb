```python
import pandas as pd
import os
import statsmodels.formula.api as sm

import regulations
import analysis
import tables
```

```python
data_path = '/Users/kylieleblancKylie/domino/dofis/data/'
table_path = '/Users/kylieleblancKylie/domino/dofis/results/'
data = pd.read_csv(os.path.join(data_path, 'clean', 'master_data.csv'),
                  sep=",")
data['cons'] = 1
```

```python
data = data[data.year == 'yr1415']
```

# Student Performance

## Schedules

```python
math_df = analysis.create_balance_df(data = data, list_of_regs = regulations.schedules,
                         reg_labels = regulations.labels, x_var = 'math')
tables.var_diff_to_excel(file = table_path + 'Student Performance.xlsx', 
                         df = math_df, 
                         control_col = 'Control', 
                         diff_col = 'Exemption Difference', 
                         se_col = 'Std. Error', pvalue_col = 'P-value',
                         start_col = 3, start_row = 5)
math_df
```

```python
reading_df = analysis.create_balance_df(data = data, list_of_regs = regulations.schedules,
                         reg_labels = regulations.labels, x_var = 'reading')
tables.var_diff_to_excel(file = table_path + 'Student Performance.xlsx', 
                         df = reading_df, 
                         control_col = 'Control', 
                         diff_col = 'Exemption Difference', 
                         se_col = 'Std. Error', pvalue_col = 'P-value',
                         start_col = 5, start_row = 5)
reading_df
```
