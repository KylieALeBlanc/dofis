```python
import sys
sys.path.append("../")

import pandas as pd
import os
import statsmodels.formula.api as sm

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

from library import regulations
from library import analysis
from library import tables

data_path = '/Users/kylieleblancKylie/domino/dofis/data/'
```

```python
data = pd.read_csv(os.path.join(data_path, 'clean', 'master_data.csv'),
                  sep=",")
data = data[data.doi == True]
regs = data[regulations.mainless]
regs = regs.drop(['reg25_0811'], axis = 1)
regs.head()
```

```python
frequent_itemsets = apriori(regs, min_support=0.07, use_colnames=True)
frequent_itemsets.head()
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules[['antecedents', 'consequents', 'confidence']].head(20)
```
