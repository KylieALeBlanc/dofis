```python
stubnames = sorted(
    set([match[0] for match in data.columns.str.findall(
    r'reg.*').values if match != [] ])
    )
stubnames
```

```python
#reg_types = [hiring, scheduling, classsize, behavior, finances]
hiring_regs = []
for reg in stubnames:
    if (reg >= 'reg21_000') & (reg < 'reg21_100'):
        hiring_regs.append(reg)
        
contract_regs = []
for reg in stubnames:
    if (reg >= 'reg21_100') & (reg < 'reg22_000'):
        contract_regs.append(reg)
        
schedule_regs = []
for reg in stubnames:
    if (reg >= 'reg25_08') & (reg < 'reg25_09'):
        schedule_regs.append(reg)
        
classsize_regs = []
for reg in stubnames:
    if (reg >= 'reg25_1') & (reg < 'reg25_2'):
        classsize_regs.append(reg)
        
behavior_regs = []
for reg in stubnames:
    if (reg >= 'reg37_000') & (reg < 'reg38_000'):
        behavior_regs.append(reg)
behavior_regs.append('reg25_036')
behavior_regs.append('reg25_092')

finance_regs = []
for reg in stubnames:
    if (reg >= 'reg44_000') & (reg < 'reg46_000'):
        finance_regs.append(reg)

```
