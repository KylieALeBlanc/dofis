```python
import pandas as pd
import os
import classify
from start import data_path
import numpy as np
import extract_laws
```

```python
labeled_docs = pd.read_csv(os.path.join(data_path, 'docs_to_label_curves_labelled.csv'))
print(labeled_docs['label'].value_counts())
labeled_docs.head(5)
labeled_docs = labeled_docs[(labeled_docs.label == True) | (labeled_docs.label == False)]
```

# Dictionary-based classifier

```python
precisions = []
recalls = []
fmeasures = []
for n in [20, 40, 60, 80, 100, 120, 140, 160, 180, 200, 220, 240, 260, 280, 300, 320, 340, 360, 380, 400]:
    sample = n
    start_ob = len(labeled_docs) - (sample + 85)
    sample_docs = labeled_docs[start_ob:]
    print('The first n-85 of documents are used for training. Training dataset statistics:')
    split = int(len(sample_docs) - 85)
    print(sample_docs.head(split)['label'].value_counts())

    print('The last 85 of documents are used for testing. Testing dataset statistics:')
    validate_docs = sample_docs.tail(85)
    print(validate_docs['label'].value_counts())

    sample_docs = sample_docs.head(split)
    len(sample_docs)
    sample_docs['laws'] = sample_docs.text.apply(extract_laws.get_laws)
    
    repeat_laws = []
    for l in sample_docs.laws:
        repeat_laws = repeat_laws + l
    laws = []
    for law in repeat_laws:
        if law not in laws:
            laws.append(str(law))
    validate_docs['keyword'] = np.where(validate_docs.text.str.contains("District of Innovation") &
                                   (validate_docs.text.str.contains('|'.join(laws))), True, False)
    tp = len(validate_docs[(validate_docs.label == True) & (validate_docs.keyword == True)])
    fp = len(validate_docs[(validate_docs.label == False) & (validate_docs.keyword == True)])
    tn = len(validate_docs[(validate_docs.label == False) & (validate_docs.keyword == False)])
    fn = len(validate_docs[(validate_docs.label == True) & (validate_docs.keyword == False)])
    print("True positives:", tp)
    print("False positives:", fp)
    print("True negatives:", tn)
    print("False negatives:", fn)
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    print("precision:", precision)
    print("recall:", recall)
    fmeasure = 2 * ((precision*recall)/(precision+recall))
    print("fmeasure:", fmeasure)
    precisions.append(precision)
    recalls.append(recall)
    fmeasures.append(fmeasure)
```
