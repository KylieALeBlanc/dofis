```python
import os
%matplotlib inline
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from start import data_path
```

```python
data = pd.read_csv(os.path.join(data_path,'doi_dates_and_exemptions_temp.csv'), sep=",")
data[['title', 'link', 'date_year', 'date_month', 'date_phrase']].head()
```

There are two ways we can determine dates - the term put forward in the plan (this is almost always just a school year) or the date a plan was finalized. We prefer the term date. However, sometimes the term is for the same AY a plan was finalized. We'll extract the term year and the month the plan was finalized, noting the few cases where these are different. 

```python
data[['title', 'term_phrase', 'term_year', 'finalize_phrase', 'finalize_year', 'finalize_month', 'link']].head()
```

### Add column called correct. Put correct year in this column. Save as fix_dates_corrected.csv. Then upload.

```python
# Incorporate previously corrected dates
fixdates_new = data[['title', 'term_phrase', 'term_year', 'finalize_phrase', 'finalize_year', 'finalize_month', 'link']]
fixdates_new = fixdates_new.rename({'term_year': 'term_year_scraped', 'finalize_year': 'finalize_year_scraped', 
                                    'finalize_month':'finalize_month_scraped'}, axis = 'columns')
fixdates_old = pd.read_csv(os.path.join(data_path, 'fix_dates_corrected.csv'), encoding = 'latin-1')
fixdates_old = fixdates_old.rename({'term_year': 'term_year_corrected', 'finalize_year': 'finalize_year_corrected',
                                   'finalize_month': 'finalize_month_corrected'}, axis = 'columns')
fixdates_old = fixdates_old[['title', 'term_year_corrected', 'finalize_year_corrected', 'finalize_month_corrected']]
fixdates = fixdates_new.merge(fixdates_old, how = 'left', left_on = 'title', right_on = 'title')
fixdates['term_year'] = fixdates['term_year_corrected']
fixdates['finalize_year'] = fixdates['finalize_year_corrected']
fixdates['finalize_month'] = fixdates['finalize_month_corrected']
fixdates['corrected'] = ''
fixdates = fixdates[['corrected', 'title', 'term_phrase', 'term_year', 'finalize_phrase', 'finalize_year', 'finalize_month', 'link']]
fixdates.to_csv(os.path.join(data_path,'fix_dates.csv'))
fixdates
```

```python
fixed_dates = pd.read_csv(os.path.join(data_path, 'fix_dates_corrected.csv'), encoding = 'latin-1')
fixed_dates.head()
```

## Set doi_year as term year if available. Otherwise, finalize year

```python
fixed_dates.loc[(pd.notnull(fixed_dates.term_year)) & (fixed_dates.term_year != -999), 'doi_year'] = fixed_dates.term_year
fixed_dates.loc[(pd.isnull(fixed_dates.term_year)) | (fixed_dates.term_year != -999), 'doi_year'] = fixed_dates.finalize_year
fixed_dates = fixed_dates.drop(columns = 'link')
fixed_dates.head(10)
```

## Drop old years from data and merge new years

```python
dates_and_exemptions = pd.read_csv(os.path.join(data_path,'doi_dates_and_exemptions_temp.csv'), sep=",")
dates_and_exemptions = dates_and_exemptions[['title', 'link', 'p_innovation', 'possible_laws', 'text']]
dates_and_exemptions = dates_and_exemptions.merge(fixed_dates, left_on = 'title', right_on = 'title', how = 'left')
```

```python
dates_and_exemptions.to_csv(os.path.join(data_path, 'doi_dates_and_exemptions.csv'))
```
